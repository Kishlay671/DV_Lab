```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df_toyota = pd.read_csv("Toyota.csv")

# Convert 'KM' and 'HP' to numeric, handling errors
df_toyota["KM"] = pd.to_numeric(df_toyota["KM"], errors="coerce")
df_toyota["HP"] = pd.to_numeric(df_toyota["HP"], errors="coerce")

# Drop rows with missing values in relevant columns
df_toyota_clean = df_toyota.dropna(subset=["Price", "Age", "KM"])

# Sample a smaller dataset to reduce clutter
df_toyota_sampled = df_toyota_clean.sample(n=100, random_state=42)

# Plot trend line with confidence band for Price vs. Age
plt.figure(figsize=(9, 6))
ax = sns.regplot(x=df_toyota_sampled["Age"], y=df_toyota_sampled["Price"], 
                 scatter_kws={'alpha':0.5}, 
                 line_kws={'color':'red'})

# Labels and title
plt.xlabel("Age (months)")
plt.ylabel("Price ()")
plt.title("Trend of Car Price vs. Car Age")

# Add legend manually
plt.scatter([], [], color='blue', alpha=0.5, label='Data Points')
plt.plot([], [], color='red', label='Trend Line')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
